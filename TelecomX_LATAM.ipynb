# Importar Librerias

```python
import pandas as pd
import numpy as np
import plotly as plt
import datetime as dt
```

#📌 Extracción

```python
TelecomX_Data = pd.read_json('TelecomX_Data.json')


```

#🔧 Transformación

## Validando datos duplicados


```python
duplicados = TelecomX_Data['customerID'].duplicated()
if duplicados.any():
   print(duplicados)
else:
   print('No hay duplicados')
```

## Ajustando el Dataframe, validando tipo de datos y ordenando.

```python
Data_customer = pd.json_normalize(TelecomX_Data['customer'])
Data_phone = pd.json_normalize(TelecomX_Data['phone'])
Data_internet = pd.json_normalize(TelecomX_Data['internet'])
Data_account = pd.json_normalize(TelecomX_Data['account'])
TelecomX_Data_concat = pd.concat([TelecomX_Data['customerID'],TelecomX_Data['Churn'],
                                  Data_customer, Data_phone, Data_internet, Data_account],
                                 axis=1)
TelecomX_Data_concat.info()
```

## Trasnformando datos en el tipo correcto según aplique.



```python
TelecomX_Data_concat['customerID'] = TelecomX_Data_concat['customerID'].astype(str)
TelecomX_Data_concat['SeniorCitizen'] = TelecomX_Data_concat['SeniorCitizen'].astype(bool)
TelecomX_Data_concat['Churn'] = TelecomX_Data_concat['Churn'].astype(bool)
TelecomX_Data_concat['Partner'] = TelecomX_Data_concat['Partner'].astype(bool)
TelecomX_Data_concat['Dependents'] = TelecomX_Data_concat['Dependents'].astype(bool)
TelecomX_Data_concat['PhoneService'] = TelecomX_Data_concat['PhoneService'].astype(bool)
TelecomX_Data_concat['MultipleLines'] = TelecomX_Data_concat['MultipleLines'].astype(bool)
TelecomX_Data_concat['OnlineBackup'] = TelecomX_Data_concat['OnlineBackup'].astype(bool)
TelecomX_Data_concat['OnlineSecurity'] = TelecomX_Data_concat['OnlineSecurity'].astype(bool)
TelecomX_Data_concat['DeviceProtection'] = TelecomX_Data_concat['DeviceProtection'].astype(bool)
TelecomX_Data_concat['TechSupport'] = TelecomX_Data_concat['TechSupport'].astype(bool)
TelecomX_Data_concat['StreamingTV'] = TelecomX_Data_concat['StreamingTV'].astype(bool)
TelecomX_Data_concat['StreamingMovies'] = TelecomX_Data_concat['StreamingMovies'].astype(bool)
TelecomX_Data_concat['PaperlessBilling'] = TelecomX_Data_concat['PaperlessBilling'].astype(bool)
TelecomX_Data_concat['gender'] = TelecomX_Data_concat['gender'].astype(str)
TelecomX_Data_concat['Contract'] = TelecomX_Data_concat['Contract'].astype(str)
TelecomX_Data_concat['PaymentMethod'] = TelecomX_Data_concat['PaymentMethod'].astype(str)
TelecomX_Data_concat['Charges.Total'] = TelecomX_Data_concat['Charges.Total'].replace(r'^\s*$', np.nan, regex=True)
TelecomX_Data_concat['Charges.Total'] = TelecomX_Data_concat['Charges.Total'].astype(float)
```

## Agregar columna "Charges.Daily"

```python
TelecomX_Data_concat['Charges.Daily'] = TelecomX_Data_concat['Charges.Monthly'] / 30.0
```

## Renombrar/Traducir Columnas

```python
TelecomX_Data_concat = TelecomX_Data_concat.rename(columns={
    'Churn': 'Cliente_Detractor',
    'gender': 'Genero',
    'SeniorCitizen': 'Usuario_Tercera_Edad',
    'Partner': 'Tiene_Pareja',
    'Dependents': 'Tiene_Dependientes',
    'tenure': 'Meses_Contrato',
    'PhoneService': 'Tiene_Servicio_Telefonico',
    'MultipleLines': 'Tiene_Lineas_multiples',
    'InternetService': 'Servicio_Internet',
    'OnlineBackup': 'Tiene_Backup_Online',
    'OnlineSecurity': 'Tiene_Seguridad_Online',
    'DeviceProtection': 'Tiene_Proteccion_Dispositivos',
    'TechSupport': 'Tiene_Soporte_Tecnico',
    'StreamingTV': 'Tiene_Streaming_TV',
    'StreamingMovies': 'Tiene_Streaming_De_Peliculas',
    'Contract': 'Duración_Contrato',
    'PaperlessBilling': 'Facturacion_Electronica',
    'PaymentMethod': 'Metodo_Pago',
    'Charges.Daily': 'Cargos_Diarios',
    'Charges.Monthly': 'Cargos_Mensuales',
    'Charges.Total': 'Total_Facturado',

})
```
