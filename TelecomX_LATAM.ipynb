# Importar Librerias

```python
import pandas as pd
import numpy as np
import plotly as plt
```

#📌 Extracción

```python
TelecomX_Data = pd.read_json('TelecomX_Data.json')

```

#🔧 Transformación

```python
duplicados = TelecomX_Data['customerID'].duplicated()
if duplicados.any():
   print(duplicados)
else:
   print('No hay duplicados')
```

```python
Data_customer = pd.json_normalize(TelecomX_Data['customer'])
Data_phone = pd.json_normalize(TelecomX_Data['phone'])
Data_internet = pd.json_normalize(TelecomX_Data['internet'])
Data_account = pd.json_normalize(TelecomX_Data['account'])
TelecomX_Data_concat = pd.concat([TelecomX_Data['customerID'],TelecomX_Data['Churn'],
                                  Data_customer, Data_phone, Data_internet, Data_account],
                                 axis=1)
```
