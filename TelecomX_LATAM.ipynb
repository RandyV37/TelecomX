## Agrupar datos

```python
# Filtrar solo los clientes detractores
detractores = TelecomX_Data_concat[TelecomX_Data_concat['Cliente_Detractor'] == True]

# Crear columna de perfil combinando Pareja y Dependientes
detractores['Perfil'] = detractores.apply(
    lambda row: f"{'Con Pareja' if row['Tiene_Pareja'] else 'Sin Pareja'} - " +
                f"{'Con Dependientes' if row['Tiene_Dependientes'] else 'Sin Dependientes'}", axis=1)

# Función auxiliar para agrupar y contar
def contar(df):
    return df.groupby('Perfil').size().reindex([
        'Con Pareja - Con Dependientes',
        'Con Pareja - Sin Dependientes',
        'Sin Pareja - Con Dependientes',
        'Sin Pareja - Sin Dependientes'
    ], fill_value=0).reset_index(name='Cantidad')

# Crear Subconjuntos
masc_tercera = contar(detractores[(detractores['Genero'] == 'Male') & (detractores['Usuario_Tercera_Edad'])])
masc_no_tercera = contar(detractores[(detractores['Genero'] == 'Male') & (~detractores['Usuario_Tercera_Edad'])])
fem_tercera = contar(detractores[(detractores['Genero'] == 'Female') & (detractores['Usuario_Tercera_Edad'])])
fem_no_tercera = contar(detractores[(detractores['Genero'] == 'Female') & (~detractores['Usuario_Tercera_Edad'])])
```
